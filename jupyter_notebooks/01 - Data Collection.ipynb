## Fetch Data from Kaggle
---
Primarily, you need to download a **JSON File (Authentication Token)** to your workspace, from Kaggle for authentication.

How to do this:
1. From the site header, click on your User Profile Picture, then on **"Account"** from the dropdown menu. This will take you to your account settings.

<p align="center">
  <img src="../README-images/kaggle-dropdown.png" width="70%"/>
</p>

2. Scroll down to the section of the page referred to as **API**

<p align="center">
  <img src="../README-images/api-token.png" width="70%"/>
</p>

3. Click **Expire API Token** to remove previous tokens.
4. To create a new token, click on the **"Create New API Token"** button. It will generate a fresh authentication token and will download **kaggle.json** file to your workspace.

<p align="center">
  <img src="../README-images/kaggle-auth.png" width="70%"/>
</p>

In case you run into any trouble, head to the **"Authentication"** section in this [link](https://www.kaggle.com/docs/api)
- Finally, you should have this file saved locally in your machine. **Please make sure this is file is named as kaggle.json**
